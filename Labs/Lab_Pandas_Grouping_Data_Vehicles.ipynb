<b>show the average CO2_Emission_Grams/Km  by Brand

```python
## Convert the CO2 Emission Grams/Mile to Emission_Grams/Km 
print(cars_df.loc[:,"CO2 Emission Grams/Mile"])
cars_df.loc[:,"CO2 Emission Grams/Mile"] = cars_df.loc[:,"CO2 Emission Grams/Mile"] / 1.60934
cars_df.rename(columns = {'CO2 Emission Grams/Mile':'CO2 Emission Grams/Km'}, inplace = True)
print(cars_df['CO2 Emission Grams/Km'].head())
```

```python
##  Show the average CO2_Emission_Grams/Km by Brand
cars_df.groupby(["Make"])["CO2 Emission Grams/Km"].mean()
```

<b>show the average CO2_Emission_Grams/Km  by Brand ... sorted

```python
cars_df.groupby(["Make"])["CO2 Emission Grams/Km"].mean().sort_values(ascending = False)

```

Use `pd.cut` or `pd.qcut` to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
cars_df['Year'].describe()
```

```python
cars_df['Year'].describe()
cars_df["Decade"] = pd.cut(cars_df["Year"],bins=[1980,1989,1999,2009,2019], labels=["80s","90s","00s","10s"], retbins=False, precision=3, duplicates='raise')
```

```python
cars_df.groupby("Decade")["Decade"].count()
```

### Did cars consume more gas in the eighties?

show the average City_Km/Liter by year_range

```python
print(cars_df.loc[:,"City MPG"])
cars_df.loc[:,"City MPG "] = cars_df.loc[:,"City MPG"] * (1.60934 / 3.78541)
cars_df.rename(columns = {'City MPG':'City KM/L'}, inplace = True)
cars_df.loc[:,"Highway MPG "] = cars_df.loc[:,"Highway MPG"] * (1.60934 / 3.78541)
cars_df.rename(columns = {'Highway MPG':'Highway KM/L'}, inplace = True)




print(cars_df['City KM/L'].head())
print(cars_df['Highway KM/L'].head())

```

```python
cars_df.groupby(['Decade'])['City KM/L'].mean()
```

Do cars with automatic transmission consume more fuel than cars with manual transmission?

```python
transmission_binary = []
for item in list(cars_df["Transmission"]):
    if item.startswith("A") == True:
        transmission_binary.append("Automatic")
    else:
        transmission_binary.append("Manual")
cars_df["Binary Transmission"] = transmission_binary
print(cars_df.head())
```

```python
cars_df
```

```python
(cars_df.groupby(["Binary Transmission"])["City KM/L"].mean()).sort_values(ascending=False)
```

Use `groupby` and `aggregate` with different aggregation measures for different columns:

aggregate with average City_Km/Liter and the count of the Trans

```python
cars_df.groupby(['Binary Transmission'])['Binary Transmission','City KM/L'].agg(['count','mean'])
```

aggregate with average City_Km/Liter and the minimum of the Trans

```python
cars_df.groupby(["Binary Transmission"])["City KM/L"].min()
```
